```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 256
N = 256
```

```python
w1 = np.pi/10  #0.314159265
w2 = np.pi/7   #0.4487989505
f1 = np.random.uniform(0, 2*np.pi)
f2 = np.random.uniform(0, 2*np.pi)
print(f1,f2)
```

```python
n = np.arange(L)
x1 = np.exp(1j*(w1*n+f1))
x2 = 0.9*np.exp(1j*(w2*n+f2))
w = 0.54 - 0.46*np.cos(2*np.pi*n/N)
y = w*(x1 + x2)
```

```python
Y = np.fft.fft(y) #DFT of y
```

```python
plt.plot(n/N,np.abs(Y),"k")
#plt.xlim(0, 100)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
w2_start = np.pi/7 #allazw onomasia tou w2 giati to xrisimopoio sto for ws "w2"
step_of_w2 = 0.001
```

```python
w2_values = np.arange(w2_start, w1-step_of_w2, -step_of_w2)
peaks = np.zeros(len(w2_values), dtype=np.complex128)
print(len(peaks))
for i, w2 in enumerate(w2_values):
    x2 = 0.9*np.exp(1j*(w2*n+f2))
    y = w*(x1 + x2)
    Y = np.fft.fft(y, N)
    if i%20==0:
        plt.figure()
        plt.title('DFT of y when w2 = %f' %w2)
        plt.plot(n/N,np.abs(Y))#np.abs
        plt.savefig("%d.pdf"%i)
    peaks[i] = Y[np.argmax(np.abs(Y))]  # Find the peak in the spectrum
#print(np.abs(peaks))
```
