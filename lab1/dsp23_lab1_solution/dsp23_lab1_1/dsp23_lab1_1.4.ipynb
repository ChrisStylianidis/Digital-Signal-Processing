```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 256
N = 1024
```

```python
w1 = np.pi*0.25   
w2 = np.pi*0.3    
#f1 = np.random.uniform(0, 2*np.pi)
#f2 = np.random.uniform(0, 2*np.pi)
f1=0.3555
f2=2.537
print(f1,f2)
```

```python
n = np.arange(L)
x1 = np.exp(1j*(w1*n+f1))
x2 = 0.05*np.exp(1j*(w2*n+f2))
w = 0.54 - 0.46*np.cos(2*np.pi*n/N) #hamming
w2 = np.ones(L) #tetragoniko
y = w*(x1 + x2)
y2 = w2*(x1 + x2)
freq=np.arange(N)
```

```python
print(len(y),len(n))
y = np.append(y, np.zeros(np.abs(N-L)))
y2 = np.append(y2, np.zeros(np.abs(N-L)))
print(len(y),len(n))
```

```python
Y = np.fft.fft(y) #DFT of y
Y2 = np.fft.fft(y2) #DFT of y
```

```python
plt.figure()
plt.plot(freq/N,np.abs(Y2),"k")
plt.xlim(0, 0.24)
plt.title("(i) Rectangle window of w[n]")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.figure()
plt.plot(freq/N,np.abs(Y),"k")
plt.xlim(0, 0.24)
plt.title("(ii) Hamming window of w[n]")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.show()
```
